```python
## Multiple Conditions for Filtering 
## Finding out people in India who have coding as a hobby and are employed 

indian_users=data[(data['Country']=="India") &(data['Employment']=="Employed, full-time") &(data['CodingActivities']=="Hobby")]
indian_users
```

```python
## Data to find users who work remote, are in India and Salary is above 100K Dollars  with languages they have workd with 

filter=data[(data['Country']=="India") &(data['RemoteWork']=="Remote") &(data['ConvertedCompYearly']>100000)]
filter.EdLevel.value_counts()
```

```python
## removing Outliers ( removing junk data to find the original maximum value)
### Salaries in US dollars , adjusted with 
max_salary=data['ConvertedCompYearly'].quantile(0.997)
max_salary
```

```python
## Outliers ( data with Junk Value) 
data[data['ConvertedCompYearly']>max_salary]
```

```python
## Most probable Highest Salary in the Data set 
data=data[data['ConvertedCompYearly']<max_salary]
data['ConvertedCompYearly'].max()
```
