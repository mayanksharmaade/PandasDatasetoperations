```python
## Improve our Model 
 ## Try different set of n_estimators
import numpy as np
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
X=heart_disease.drop("target",axis=1)

##Create Y ( label Matrix)

Y=heart_disease["target"]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
clf.fit(X_train,Y_train);

np.random.seed(42)
for i in range(10,100,10):
    print(f"trying model with {i} n_estimators")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train,Y_train)
    print(f"Test accuracy on model data : {clf.score(X_test,Y_test)*100:.2f}%") 

    # Save model 
    import pickle
    pickle.dump(clf, open("random_forest_model1","wb"))
    
```

```python
loaded_model=pickle.load(open("random_forest_model1","rb"))
loaded_model.score(X_test,Y_test)
```
