```python
import bs4
```

```python
url="https://www.geeksforgeeks.org/python/python-programming-language-tutorial/#39"
```

```python
res=requests.get(url)
res.status_code
```

```python
res.text
from bs4 import BeautifulSoup 
```

```python
# Parsing the HTML
soup = BeautifulSoup(res.content, 'html.parser')
print(soup.prettify())
```

```python
# Getting the name of the tag
print(soup.title.name)

# Getting the name of parent tag
print(soup.title.parent.name)
```

```python
s = soup.find('div', class_='mainSubHeaderDiv')
print(s)
```

```python
links = s.find_all('li')
print(links)
```

```python
tags=soup.find('a', class_='link')
print(tags)
```

```python
text="Data Science"
find= soup.find_all(lambda tag: tag.name == "a" and text in tag.text)
print(find)

    
```
